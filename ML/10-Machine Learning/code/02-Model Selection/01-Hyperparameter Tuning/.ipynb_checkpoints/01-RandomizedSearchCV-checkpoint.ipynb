```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
df = pd.read_csv('advertising.csv')
df
```

```python
df = pd.get_dummies(df, columns=['City', 'Country'], drop_first=True)
df
```

```python
x = df.drop(['Ad Topic Line', 'Timestamp', 'Clicked on Ad'], axis=1)
y = df['Clicked on Ad']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()

scaler.fit(x_train)

x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(random_state=42)
```

```python
model = RandomForestClassifier(random_state=42)
model.fit(x, y)
```

```python
y_pred = model.predict(x_test)
print(confusion_matrix(y_test, y_pred))
print(model.score(x_test, y_test))
```

```python

params = [
        {'criterion':['gini', 'entropy'], 'n_estimators':range(20, 500)}
         ]

randomized_search = RandomizedSearchCV(estimator=model,
                           param_distributions=params,
                           scoring='accuracy',
                           cv=10, n_jobs=-1)
randomized_search.fit(x, y)
```

```python
randomized_search.best_score_
```

```python
randomized_search.best_params_
```

```python
model = randomized_search.best_estimator_
```
